```python
import streamlit as st
import numpy as np
import pickle

# Load pipeline
pipeline = pickle.load(open("pipeline.pkl", "rb"))

# Page config
st.set_page_config(page_title="ML Prediction App", layout="centered")

# Title
st.title("🧠 Machine Learning Prediction App")
st.markdown("Enter the details below to get prediction")

# ---------------------------
# User Inputs (EDIT THESE AS PER YOUR FEATURES)
# ---------------------------

gender = st.selectbox("Gender", [0, 1])
partner = st.selectbox("Partner", [0, 1])
dependents = st.selectbox("Dependents", [0, 1])
tenure = st.number_input("Tenure", min_value=0)
monthly_charges = st.number_input("Monthly Charges", min_value=0.0)
total_charges = st.number_input("Total Charges", min_value=0.0)

# Add remaining features if your model expects them
# Make sure COUNT matches training features after encoding

# ---------------------------
# Prediction Button
# ---------------------------
if st.button("Predict"):

    # Convert input to numpy array (MUST MATCH training feature order)
    features = np.array([[gender, partner, dependents, tenure, monthly_charges, total_charges]])

    # Prediction
    prediction = pipeline.predict(features)[0]
    probability = pipeline.predict_proba(features)[0][1]

    # Output
    st.subheader("📊 Result")

    if prediction == 1:
        st.error(f"Prediction: Positive 🚨")
    else:
        st.success(f"Prediction: Negative ✅")

    st.write(f"Confidence: {probability:.2f}")
```

```python
import streamlit as st
import numpy as np
import pickle
import plotly.graph_objects as go
import plotly.express as px

# Load artifacts
model = pickle.load(open("model.pkl", "rb"))
scaler = pickle.load(open("scaler.pkl", "rb"))
pca = pickle.load(open("pca.pkl", "rb"))

# Page config (dark feel)
st.set_page_config(page_title="Churn Predictor", layout="wide")

# Custom CSS (modern UI styling)
st.markdown("""
<style>
body {
    background-color: #0e1117;
    color: white;
}
.card {
    background-color: #1c1f26;
    padding: 20px;
    border-radius: 12px;
    box-shadow: 2px 2px 10px rgba(0,0,0,0.3);
}
</style>
""", unsafe_allow_html=True)

# Title
st.title("📊 Customer Churn Prediction Dashboard")
st.markdown("Predict churn with an interactive AI-powered system")

st.divider()

# Sidebar Inputs
st.sidebar.header("Customer Details")

gender = st.sidebar.selectbox("Gender", ["Male", "Female"])
partner = st.sidebar.selectbox("Partner", ["Yes", "No"])
dependents = st.sidebar.selectbox("Dependents", ["Yes", "No"])

contract = st.sidebar.selectbox(
    "Contract",
    ["Month-to-month", "One year", "Two year"]
)

payment = st.sidebar.selectbox(
    "Payment Method",
    ["Electronic check", "Mailed check", "Bank transfer", "Credit card"]
)

internet = st.sidebar.selectbox(
    "Internet Service",
    ["DSL", "Fiber optic", "No"]
)

tenure = st.sidebar.slider("Tenure", 0, 72, 12)
monthly = st.sidebar.number_input("Monthly Charges", 0.0, 200.0, 70.0)
total = st.sidebar.number_input("Total Charges", 0.0, 10000.0, 1000.0)

# Encoding
gender = 1 if gender == "Male" else 0
partner = 1 if partner == "Yes" else 0
dependents = 1 if dependents == "Yes" else 0

contract_map = {"Month-to-month": 0, "One year": 1, "Two year": 2}
payment_map = {
    "Electronic check": 0,
    "Mailed check": 1,
    "Bank transfer": 2,
    "Credit card": 3
}
internet_map = {"DSL": 0, "Fiber optic": 1, "No": 2}

contract = contract_map[contract]
payment = payment_map[payment]
internet = internet_map[internet]

# Feature vector
features = np.array([[
    gender,
    partner,
    dependents,
    contract,
    payment,
    internet,
    tenure,
    monthly,
    total
]])

# Preprocessing
features_scaled = scaler.transform(features)
features_pca = pca.transform(features_scaled)

# Prediction
if st.button("🚀 Predict Churn"):

    pred = model.predict(features_pca)[0]
    prob = model.predict_proba(features_pca)[0][1]

    col1, col2 = st.columns(2)

    with col1:
        st.markdown('<div class="card">', unsafe_allow_html=True)
        st.subheader("📥 Input Summary")
        st.write(features)
        st.markdown('</div>', unsafe_allow_html=True)

    with col2:
        st.markdown('<div class="card">', unsafe_allow_html=True)
        st.subheader("🔮 Prediction Result")

        if pred == 1:
            st.error("⚠️ Customer is likely to CHURN")
        else:
            st.success("✅ Customer is likely to STAY")

        st.write(f"Probability: {prob:.2f}")
        st.markdown('</div>', unsafe_allow_html=True)

    st.divider()

    # =========================
    # 📊 Probability Gauge Chart
    # =========================
    fig = go.Figure(go.Indicator(
        mode="gauge+number",
        value=prob * 100,
        title={'text': "Churn Probability (%)"},
        gauge={
            'axis': {'range': [0, 100]},
            'bar': {'color': "red"},
            'steps': [
                {'range': [0, 40], 'color': "green"},
                {'range': [40, 70], 'color': "yellow"},
                {'range': [70, 100], 'color': "red"}
            ],
        }
    ))

    st.plotly_chart(fig, use_container_width=True)

    # =========================
    # 📈 Feature Importance (if available)
    # =========================
    try:
        importance = model.feature_importances_

        feature_names = [
            "gender","partner","dependents","contract",
            "payment","internet","tenure","monthly","total"
        ]

        fig2 = px.bar(
            x=importance,
            y=feature_names,
            orientation='h',
            title="Feature Importance"
        )

        st.plotly_chart(fig2, use_container_width=True)

    except:
        st.info("Feature importance not available for this model.")
```
